```python
topics = ['Cricket Tournament','Boris Johnson','NHS service','Tory leadership','Cancer treatment','Formula One','Lewis Hamilton','Russia Ukraine',
          'Commonwealth Games','Covid-19','Wimbledon Championship','Monkey pox','Labour party','Quarantine rules', 'mental health', 'Euro 2022',
          'Harry Kane Euro 2022','Sri Lankan economic crisis','health insurance','hepatitis','ADHD','Tour de France','Sebastian Vettel formula one',
          'Joe Biden US','Fertility Treatment','Shinzo Abe','Physical Disability',' Copa America Femenina','Fernando Alonso formula one',
          'Fidel Ramos','Scottish Referendum','Betsi Cadwaladr', 'wheelchair marathon', 'Liz Truss and Rishi Sunak','Brexit',
          'Exercise', 'heart gene therapy', 'Dementia',' Victor Kiplangat','World Health Organisation', 'VAccination','Acid reflux', 'Gymnastics',
          'G7 international tax','Diabetes','Liverpool vs Manchester City','Pandora papers leaked documents','Rugby League','New Zealand Fuel Price',
          'Washington Open','Conservative party','Angela Merkel','AIDS in children', 'Global Influenza Surveillance and Response System ','Chloe Kelly', 'World Health Day',
          'FIFA world cup','Swimmer Anita Alvarez','Novak Djokovic','Womens world cup','Brain Tumor Treatment','Fernondo Alonso',
          'Alzheimers Disease','Myanmar Crisis','Abortion rights', 'Peoples Republic of China ','Centaurus','T20 World Cup',
          'Alder Hey','SuperMotocross World Championship','Yoga','Food Diet','Premier League Football','kuwait Parliament',
          'Democratic party','Pelosi Asian Tour','Chess','Breastfeeding','Marburg Virus','Cristiano Ronaldo',
          'Chickenpox','Boxing','Peter Meijer house republican','Body image and mental health','Chelsea vs Barcelona',
          'paxlovid','Hong Kong Sinovac Vaccine','Donald Trump','Championship League Snooker',
          'NATO','Monkey pox vaccine','Football association challenge cup','Conjoined twins separated',
          'Priti Patel Rwanda deportation plan','Covid booster',
          'doping in sports','Nikita Parris from WSL ','Polio virus','Tiger Woods Golf',' Election Results']
```

```python
import wikipedia
docs =[]
for article in topics:
 
# finding result for the search
# sentences = 10 refers to numbers of line
    result = wikipedia.summary(article, sentences = 10)
    docs.append(result)
 
# printing the result
#print(docs)
```

```python
print(docs)
```

```python
from nltk.corpus import stopwords
sw = stopwords.words('english')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

ps = PorterStemmer()
filtered_docs = []
for doc in docs:
    tokens = word_tokenize(doc)
    tmp = ""
    for w in tokens:
        if w not in sw:
            tmp += ps.stem(w) + " "
    filtered_docs.append(tmp)
print(filtered_docs)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer =TfidfVectorizer()
X = vectorizer.fit_transform(filtered_docs)
print(X.todense())
```

```python
from sklearn.cluster import KMeans
K = 3 
model = KMeans(n_clusters=K)#, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("cluster no. of input documents, in the order they received:")
print(model.labels_)
```

```python
test_doc = ["I play basketball",
             "Union strike against the government",
             "The world is under a COVID pandemic"
           ]
filtered_test_docs = []
for doc in test_doc:
    tokens = word_tokenize(doc)
    tmp = ""
    for w in tokens:
        if w not in sw:
            tmp += ps.stem(w) + " "
    filtered_test_docs.append(tmp)
    
print(filtered_test_docs)

Y = vectorizer.transform([filtered_test_docs[0]])
prediction = model.predict(Y)
print(prediction)

Y = vectorizer.transform([filtered_test_docs[1]])
prediction = model.predict(Y)
print(prediction)

Y = vectorizer.transform([filtered_test_docs[2]])
prediction = model.predict(Y)
print(prediction)
```
