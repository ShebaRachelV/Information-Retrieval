```python
from bs4 import BeautifulSoup
import requests

url = "https://pureportal.coventry.ac.uk/en/organisations/school-of-economics-finance-and-accounting/publications/"
page = requests.get(url)
print(page.text)
```

```python
soup = BeautifulSoup(page.text, "html.parser") 
```

```python
import time
page = 0
research_outputs = []
#to go through all the pages from 1 to 13
while page != 13:
      url = f"https://pureportal.coventry.ac.uk/en/organisations/school-of-economics-finance-and-accounting/publications/?page={page}"
      response = requests.get(url)
      html = response.content
      soup = BeautifulSoup(html, "lxml")
      for output in soup.find_all("div", class_="result-container"):
            research_output = {}
            research_output["title"] = output.h3.get_text(strip=True)
            research_output["title_link"] = output.find("a", class_ = "link")['href']
            research_output["year"] = output.find("span", class_="date").get_text()[-4:]
            link = output.find("a", class_ = "link")['href']
            page_1 = requests.get(link)
            soup_1 = BeautifulSoup(page_1.content, features = "lxml")
            abstract = soup_1.find("div",class_ = "textblock")
            if abstract is not None:
                research_output["abstract"]=abstract.get_text(strip=True)
            research_output["authors_name"] = soup_1.find("p", class_="relations persons").get_text(strip=True)
            for author in soup_1.find_all("a", class_="link person"):
                if author is not None:
                    research_output["author_link"] = author.get('href')
                    research_output["author_cov"] = author.get_text(strip=True)
            research_outputs.append(research_output)
      page = page + 1
      time.sleep(0.5)
```

```python
import pandas as pd
res_df = pd.DataFrame(research_outputs)
res_df
```

```python
res_df.to_csv('research_output.csv',index= False)
```

```python
import pandas as pd
research = pd.read_csv("research_output.csv")
research
```

```python
res = research.fillna('abc')
```

```python

from nltk.corpus import stopwords

sw = stopwords.words('english')
sw.append('abc')
print(sw)
```

```python


from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

r = res['title']+ ' ' + res['authors_name'] + ' ' + (res['year']).astype(str) + ' ' + res['abstract']
ps = PorterStemmer()
filtered_docs = []

for i in range(0,637):
    tokens = word_tokenize(r[i])
    tmp = ""
    for w in tokens:
        if w not in sw:
            tmp += ps.stem(w) + " "
    filtered_docs.append(tmp)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Instantiate a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# It fits the data and transform it as a vector
X = vectorizer.fit_transform(filtered_docs)
# Convert the X as transposed matrix
X = X.T.toarray()
#print(X.todense())
# Create a DataFrame and set the vocabulary as the index
df = pd.DataFrame(X, index=vectorizer.get_feature_names_out())
df
```

```python
import numpy as np
def get_similar_articles(q, df):
  print("query:", q)
  ps = PorterStemmer()
  filtered_query = []
  tokens = word_tokenize(q)
  tmp = ""
  for w in tokens:
        if w not in sw:
            tmp += ps.stem(w) + " "
  filtered_query.append(tmp)
            
  print("Article with the highest cosine similarity value: ")
  # Convert the query become a vector
  
  q_vec = vectorizer.transform(filtered_query).toarray().reshape(df.shape[0],)
  sim = {}
  # Calculate the similarity
  for i in range(635):
    sim[i] = np.dot(df.loc[:, i].values, q_vec) / (np.linalg.norm(df.loc[:, i]) * np.linalg.norm(q_vec))

  # Sort the values 
  sim_sorted = sorted(sim.items(), key=lambda x: x[1], reverse=True)
  #print(sim_sorted)
  # Print the articles and their similarity values
  for k,v in sim_sorted:
        print("Similarity Value:", v)
        print(research["title"][k],"(",research["year"][k],")")
        print(research["title_link"][k])
        print(research["authors_name"][k])
        print(research["author_cov"][k],research["author_link"][k])
        print()
        
# Add The Query
q1 = 'financial crisis'
get_similar_articles(q1, df)
```
